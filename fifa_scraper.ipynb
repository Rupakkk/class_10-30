```python
import requests
from bs4 import BeautifulSoup
url=BeautifulSoup('https://www.fifa.com/fifa-world-ranking/ranking-table/men/#all','html.parser')
response=requests.get(url)
soup=response.text
soup=BeautifulSoup(soup,'lxml')
table=soup.table
tags=table.find_all('tr')
data=[]
for i in tags:
    data.append(i.text)
# print(data[0])
# print(len(data[1])      
final_output=[]
for i in range(len(data)):
    x=data[i].split('\n')[1:-1]
    final_output.append(x)   
finals=[]
ls=[0,7,13,14,15]
for i in final_output:
    jj=[]
    for j in ls:
        jj.append(i[j])
    finals.append(jj)
finals[0]=['RANK','TEAM','TOTAL POINTS','PREVIOUS POINTS','+/-']    
finals
        
    
import csv
with open('fifa.csv','w') as f:
    for i in finals:
        x=csv.writer(f)
        x.writerow(i)
import pandas as pd
df=pd.read_csv('fifa.csv',encoding='latin',index_col='RANK')
df
    
```
